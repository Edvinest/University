```python
from random import random
from random import choice
from time import time

#lab 8

#Újrahasználható függvények
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

#1. feladat
def MillerRabin(nr, t=10):
    r,s=nr-1,0
    while r%2==0:
        r=r//2
        s+=1
    for i in range (t):
        x= randint(2, nr)
        temp=pow(x,r,nr)
        if temp == 1 or temp==-1:
            continue
        for j in range(0,s):
            temp=(temp*temp)%nr
            if temp==nr-1:
                break
            if temp==-1:
                return False
        if temp!=nr-1:
            return False
    return True

#2. feladat

def find_primes_in_file(input_file, output_file):
    try:
        with open(input_file, "r") as infile, open(output_file, "w") as outfile:
            for line in infile:
                try:
                    numbers = map(int, line.split())
                    for number in numbers:
                        if is_prime(number):
                            outfile.write(f"{number}\n")
                except ValueError:
                    print(f"Warning: Non-numeric data found in the line: {line.strip()}")
        
        print(f"Prime numbers have been written to {output_file}.")
    except FileNotFoundError:
        print(f"Error: File '{input_file}' not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

#3. feladat
def is_safe_prime(n):
    if n < 2 or (n > 2 and n % 2 == 0):
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    
    m = (n - 1) // 2
    if m < 2:
        return False
    i = 3
    while i * i <= m:
        if m % i == 0:
            return False
        i += 2
    return True

#4. feladat
def find_generators_for_safe_prime(p):
    if not is_prime(p):
        raise ValueError("Input must be a prime number.")
    
    generators = []
    phi = p - 1
    
    for a in range(1, p):
        powers = set()
        is_generator = True
        for k in range(1, phi + 1):
            mod_result = pow(a, k, p)
            if mod_result in powers:
                is_generator = False
                break
            powers.add(mod_result)
        
        if is_generator:
            generators.append(a)
    
    return generators

#5. feladat
def generate_safe_prime(n):
    while True:
        p = random.getrandbits(n) | (1 << (n - 1)) | 1
        if is_prime(p) and is_prime((p - 1) // 2):
            return p

def find_generator(p):
    phi = p - 1
    q = phi // 2
    for g in range(2, p):
        if pow(g, 2, p) != 1 and pow(g, q, p) != 1:
            return g
    return None

def to_base256(num):
    byte_array = []
    while num > 0:
        byte_array.append(num % 256)
        num //= 256
    return bytes(reversed(byte_array))

def from_base256(byte_array):
    num = 0
    for byte in byte_array:
        num = num * 256 + byte
    return num

def to_base64(byte_array):
    return base64.b64encode(byte_array).decode('utf-8')

def from_base64(encoded_str):
    return base64.b64decode(encoded_str)

def generate_and_save_primes(file_name, num_primes, bit_range):
    with open(file_name, 'w') as f:
        for _ in range(num_primes):
            n = random.randint(*bit_range)
            p = generate_safe_prime(n)
            g = find_generator(p)
            
            p_bytes = to_base256(p)
            g_bytes = to_base256(g)
            
            p_b64 = to_base64(p_bytes)
            g_b64 = to_base64(g_bytes)
            
            f.write(f"{p_b64} {g_b64}\n")

def read_and_decode_primes(file_name):
    results = []
    with open(file_name, 'r') as f:
        for line in f:
            p_b64, g_b64 = line.strip().split()
            p_bytes = from_base64(p_b64)
            g_bytes = from_base64(g_b64)
            
            p = from_base256(p_bytes)
            g = from_base256(g_bytes)
            
            results.append((p, g))
    return results

#6. feladat
def diffie_hellman(p, g):
    # Privát kulcsok
    a = random.randint(2, p - 2)
    b = random.randint(2, p - 2)
    
    # Nyilvános kulcsok
    A = pow(g, a, p)
    B = pow(g, b, p)
    
    # Közös titok
    shared_secret_A = pow(B, a, p)
    shared_secret_B = pow(A, b, p)
    
    assert shared_secret_A == shared_secret_B, "A közös titoknak egyeznie kell"
    return shared_secret_A


        
def main():
    klen=32
    key=bytes([])
    for i in range(klen):
        key+=bytes([randint(256)])
        
    encrypt(key)
    print("Sikerült titkosítani")
```
