```python
import random

#Generáljunk véletlenszerűen egy pontosan n-bites számot és határozzuk meg a bináris alakját.

def nBits(n):
    b1 = 1 << (n-1)
    b2 = (1 << n) - 1
    nr = random.randint(b1, b2)
    return nr
    
def countDiff(number):
    number_str = format(number, 'b')
    ones = number_str.count('1')
    zeros = number_str.count('0')
    return ones - zeros

def bitwiseOps(n, num_pairs):
    result = []
    for _ in range(num_pairs):
        num1 = nBits(n)
        num2 = nBits(n)
        
        and_result = num1 & num2
        or_result = num1 | num2
        xor_result = num1 ^ num2
        
        and_difference = countDiff(and_result);
        or_difference = countDiff(or_result);
        xor_difference = countDiff(xor_result);
        
        result.append({
            "num1": num1,
            "num2": num2,
            "and_diff": and_difference,
            "or_diff": or_difference,
            "xor_diff": xor_difference
        }
        )
        
    return result
    

results = bitwiseOps(8, 5)

for i, result in enumerate(results):
    print(f"Pair {i+1}")
    print(f"Number 1: {result['num1']} ({format(result['num1'], '08b')})")
    print(f"Number 2: {result['num2']} ({format(result['num2'], '08b')})")
    print(f"AND difference: {result['and_diff']}")
    print(f"OR difference: {result['or_diff']}")
    print(f"XOR difference: {result['xor_diff']}")
    print()
```

```python
#Írjunk programot, amely egy bináris állomány bájtjainak megfelelő hexa, illetve bináris értékeit kiírja egy szövegállományba. 
#A hexa formátum esetében minden értéket két szimbólummal és egy space-el, bináris formátum esetén minden értéket 8 szimbólummal 
#és egy space-el jelenítsünk meg. Egy sorba hexa formátum esetén 16 értéket, a bináris formátum esetén pedig 4 értéket tegyünk

def convert_file_to_hex_and_bin(input_file, hex_output_file, bin_output_file):
    try:
        with open(input_file, "rb") as bin_file:
            data = bin_file.read()
        
        with open(hex_output_file, "w") as hex_file:
            for i, byte in enumerate(data):
                hex_file.write(f"{byte:02X} ")
                if (i + 1) % 16 == 0:
                    hex_file.write("\n")
            if len(data) % 16 != 0:
                hex_file.write("\n")


        with open(bin_output_file, "w") as bin_file:
            for i, byte in enumerate(data):
                bin_file.write(f"{byte:08b} ")
                if (i + 1) % 4 == 0:
                    bin_file.write("\n")
            if len(data) % 4 != 0:
                bin_file.write("\n")

        print(f"Hexadecimal output written to {hex_output_file}")
        print(f"Binary output written to {bin_output_file}")
    except FileNotFoundError:
        print(f"Error: File '{input_file}' not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
