```python
# 1. feladat

# Generáljunk véletlenszerűen egy-egy n bites számot és határozzuk 
# meg a két szám bitenkénti AND, OR, XOR értékeit, majd határozzuk meg, 
# hogy a kapott eredményekben, mennyi a különbség az 1-es és 0-ás bitek 
# száma között. Több számpárra is végezzük el a vizsgálatot! Mit veszünk észre?

import random

# Véletlenszám generálás n bit hosszúságú számokhoz
def generate_random_number(n):
    return random.randint(0, (1 << n) - 1)

# Különbség számítása a 1-es és 0-ás bitek száma között
def count_ones_and_zeros(num):
    # Bináris formátumban a szám, kivágva a '0b' előtagot
    bin_repr = bin(num)[2:]
    ones_count = bin_repr.count('1')
    zeros_count = len(bin_repr) - ones_count
    return ones_count, zeros_count

# Műveletek végrehajtása két számra
def calculate_bitwise_operations(a, b):
    and_result = a & b
    or_result = a | b
    xor_result = a ^ b
    return and_result, or_result, xor_result

n = int(input("Adja meg a szám bit hosszát (n): "))
pairs_count = int(input("Hány véletlenszám párra végezze el a vizsgálatot? "))

for _ in range(pairs_count):
    # Véletlenszámok generálása
    a = generate_random_number(n)
    b = generate_random_number(n)
        
    # Műveletek végrehajtása
    and_result, or_result, xor_result = calculate_bitwise_operations(a, b)
        
    # A bitenkénti műveletek eredményeinek számlálása
    and_ones, and_zeros = count_ones_and_zeros(and_result)
    or_ones, or_zeros = count_ones_and_zeros(or_result)
    xor_ones, xor_zeros = count_ones_and_zeros(xor_result)
        
    # Eredmények kiíratása
    print(f"\nA számok: {a} ({bin(a)[2:].zfill(n)}), {b} ({bin(b)[2:].zfill(n)})")
    print(f"AND művelet eredménye: {and_result} ({bin(and_result)[2:].zfill(n)})")
    print(f"  1-esek száma: {and_ones}, 0-ásak száma: {and_zeros}")
        
    print(f"OR művelet eredménye: {or_result} ({bin(or_result)[2:].zfill(n)})")
    print(f"  1-esek száma: {or_ones}, 0-ásak száma: {or_zeros}")
        
    print(f"XOR művelet eredménye: {xor_result} ({bin(xor_result)[2:].zfill(n)})")
    print(f"  1-esek száma: {xor_ones}, 0-ásak száma: {xor_zeros}")
        
    print(f"\nKülönbség (1-esek - 0-ásak) az eredményekben:")
    print(f"AND: {and_ones - and_zeros}")
    print(f"OR: {or_ones - or_zeros}")
    print(f"XOR: {xor_ones - xor_zeros}")
```

```python
# 2. feladat

# A tanult két módszerrel határozzuk meg, hogy hány 1-es bit van 2^n, 2^{n-1}, 2^{n+1} 
# számokban, illetve egy véletlenszerűen generált n-bites számban, ahol n >= 106. 
# Mérjük le a számításokhoz szükséges futási időt.

import random
import time

# Bináris reprezentáció módszerrel
def count_ones_binary_method(num):
    return bin(num).count('1')

# Bitwise módszerrel
def count_ones_bitwise_method(num):
    count = 0
    while num:
        count += num & 1  # Ellenőrizzük az aktuális bitet
        num >>= 1  # Jobbra toljuk a számot egy hellyel
    return count

    # N értékének beállítása
    n = int(input("Adja meg a kívánt n értékét (n >= 106): "))
    
    # Ha n < 106, kérjük, hogy a felhasználó adjon meg nagyobbat
    if n < 106:
        print("n értékének legalább 106-nak kell lennie.")
        return
    
# Számok: 2^n, 2^(n-1), 2^(n+1) és egy véletlenszerű n-bites szám
num_2n = 2 ** n
num_2nm1 = 2 ** (n - 1)
num_2np1 = 2 ** (n + 1)
random_num = random.getrandbits(n)  # Véletlenszerű n-bites szám
    
# Számítások a bináris módszerrel
start_time = time.time()
count_ones_2n_bin = count_ones_binary_method(num_2n)
count_ones_2nm1_bin = count_ones_binary_method(num_2nm1)
count_ones_2np1_bin = count_ones_binary_method(num_2np1)
count_ones_random_bin = count_ones_binary_method(random_num)
end_time = time.time()
binary_method_time = end_time - start_time
    
# Számítások a bitwise módszerrel
start_time = time.time()
count_ones_2n_bitwise = count_ones_bitwise_method(num_2n)
count_ones_2nm1_bitwise = count_ones_bitwise_method(num_2nm1)
count_ones_2np1_bitwise = count_ones_bitwise_method(num_2np1)
count_ones_random_bitwise = count_ones_bitwise_method(random_num)
end_time = time.time()
bitwise_method_time = end_time - start_time
    
# Eredmények kiírása
print(f"\nBináris módszer eredményei:")
print(f"2^{n} szám 1-es bitek száma: {count_ones_2n_bin}")
print(f"2^{n-1} szám 1-es bitek száma: {count_ones_2nm1_bin}")
print(f"2^{n+1} szám 1-es bitek száma: {count_ones_2np1_bin}")
print(f"Véletlenszerű {n}-bites szám 1-es bitek száma: {count_ones_random_bin}")
print(f"Bináris módszer futási ideje: {binary_method_time:.6f} másodperc")
    
print(f"\nBitwise módszer eredményei:")
print(f"2^{n} szám 1-es bitek száma: {count_ones_2n_bitwise}")
print(f"2^{n-1} szám 1-es bitek száma: {count_ones_2nm1_bitwise}")
print(f"2^{n+1} szám 1-es bitek száma: {count_ones_2np1_bitwise}")
print(f"Véletlenszerű {n}-bites szám 1-es bitek száma: {count_ones_random_bitwise}")
print(f"Bitwise módszer futási ideje: {bitwise_method_time:.6f} másodperc")
```

```python
# 3. feladat

# Írjunk programot, amely egy bináris állomány bájtjainak megfelelő hexa, 
# illetve bináris értékeit kiírja egy szövegállományba. A hexa formátum esetében 
# minden értéket két szimbólummal és egy space-el, bináris formátum esetén minden 
# értéket 8 szimbólummal és egy space-el jelenítsünk meg. Egy sorba hexa formátum 
# esetén 16 értéket, a bináris formátum esetén pedig 4 értéket tegyünk.

def byte_to_hex(byte):
    """Átalakítja a bájt értéket hexa formátumra."""
    return f"{byte:02X}"

def byte_to_bin(byte):
    """Átalakítja a bájt értéket bináris formátumra."""
    return f"{byte:08b}"

def write_hex_and_bin(input_file, output_file):
    """Beolvassa a bináris fájlt, majd kiírja a hexa és bináris értékeket a szövegfájlba."""
    with open(input_file, "rb") as f_in, open(output_file, "w") as f_out:
        # A fájl bájtjainak beolvasása
        bytes_data = f_in.read()
        
        # Hexa formátumban történő kiírás
        f_out.write("Hexa Format:\n")
        hex_count = 0
        for byte in bytes_data:
            # Hexa formátumban kiírás
            f_out.write(byte_to_hex(byte) + " ")
            hex_count += 1
            if hex_count == 16:  # Minden sor végén új sort kezdünk
                f_out.write("\n")
                hex_count = 0
        
        # Bináris formátumban történő kiírás
        f_out.write("\nBinary Format:\n")
        bin_count = 0
        for byte in bytes_data:
            # Bináris formátumban kiírás
            f_out.write(byte_to_bin(byte) + " ")
            bin_count += 1
            if bin_count == 4:  # Minden sor végén új sort kezdünk
                f_out.write("\n")
                bin_count = 0

# Fájlok megadása
input_file = "input.bin"  # A bemeneti bináris fájl
output_file = "output.txt"  # A kimeneti szövegfájl, ahol a formázott értékek lesznek

# Függvény meghívása
write_hex_and_bin(input_file, output_file)
```

```python
# 4. feladat

# A 3. feladatnál létrehozott szövegállományt alakítsuk vissza a megfelelő bináris állománnyá.

def hex_to_byte(hex_str):
    """Hexadecimális formátumból bájtba alakít."""
    return bytes.fromhex(hex_str)

def bin_to_byte(bin_str):
    """Bináris formátumból bájtba alakít."""
    return int(bin_str, 2).to_bytes((len(bin_str) + 7) // 8, byteorder='big')

def convert_to_binary(input_file, output_file):
    """Visszaalakítja a hexadecimális és bináris adatokat bináris fájlba."""
    with open(input_file, "r") as f_in, open(output_file, "wb") as f_out:
        lines = f_in.readlines()
        
        # Változók a hexadecimális és bináris adataink tárolására
        hex_data = []
        bin_data = []

        for line in lines:
            # Ha a sor hexadecimális formátumban van
            if line.startswith("Hexa Format:"):
                continue  # Ugrunk a következő sorra, ha ez a fejléc
            elif line.startswith("Binary Format:"):
                continue  # Ugrunk a következő sorra, ha ez a fejléc
            else:
                # Sorok feldolgozása
                hex_values = line.split()
                bin_values = line.split()
                
                for value in hex_values:
                    if value.strip():  # Üres értékek ellenőrzése
                        hex_data.append(value.strip())
                
                for value in bin_values:
                    if value.strip():  # Üres értékek ellenőrzése
                        bin_data.append(value.strip())
        
        # Hexadecimális értékek visszaalakítása bájtokká
        for hex_val in hex_data:
            f_out.write(hex_to_byte(hex_val))
        
        # Bináris értékek visszaalakítása bájtokká
        for bin_val in bin_data:
            f_out.write(bin_to_byte(bin_val))
        
        print("A fájl sikeresen vissza lett alakítva!")

# Fájlok megadása
input_file = "output.txt"  # Az előzőleg generált szövegfájl
output_file = "reconstructed_output.bin"  # Az új bináris fájl, ahová a visszaalakított adatokat írjuk

# Függvény meghívása
convert_to_binary(input_file, output_file)
```

```python
# 5. feladat

# Írjuk ki egy szövegállományba, formázva a nyomtatható ASCII karaktereket.
def is_printable_ascii(byte):
    """Ellenőrzi, hogy a byte érték egy nyomtatható ASCII karakter-e."""
    return 32 <= byte <= 126  # Az ASCII nyomtatható karakterek tartománya

def write_printable_ascii(input_file, output_file):
    """Kiírja a nyomtatható ASCII karaktereket a bináris fájlból egy szövegfájlba."""
    with open(input_file, "rb") as bin_file, open(output_file, "w") as text_file:
        byte = bin_file.read(1)
        while byte:
            byte_value = ord(byte)  # A byte értékének lekérése
            if is_printable_ascii(byte_value):
                text_file.write(chr(byte_value))  # A byte karakterként történő írása
            byte = bin_file.read(1)  # Következő byte beolvasása

    print(f"A nyomtatható ASCII karakterek sikeresen ki lettek írva a {output_file} fájlba!")

# Fájlok megadása
input_file = "input_binary_file.bin"  # A bemeneti bináris fájl
output_file = "output_text_file.txt"  # A kimeneti szövegfájl

# Függvény meghívása
write_printable_ascii(input_file, output_file)
```

```python
# 6. feladat

# Nagybetűsítsük egy szövegállomány tartalmát, azaz minden kis angol ábécébeli 
# betűt alakítsunk át nagybetűvé a többit, hagyjuk változatlanul.

def convert_to_uppercase(input_file, output_file):
    """Beolvassa az input fájlt, nagybetűsre alakítja a szöveget, és kiírja az output fájlba."""
    with open(input_file, 'r') as file:
        text = file.read()  # A fájl teljes tartalmának beolvasása

    # A szöveg nagybetűsítése
    uppercased_text = text.upper()

    # Kiírjuk a nagybetűs szöveget az output fájlba
    with open(output_file, 'w') as file:
        file.write(uppercased_text)

    print(f"A szöveg sikeresen nagybetűsítve lett és ki lett írva a {output_file} fájlba!")

# Fájlok megadása
input_file = "input_text_file.txt"  # Bemeneti szövegfájl
output_file = "output_text_file.txt"  # Kimeneti szövegfájl

# Függvény meghívása
convert_to_uppercase(input_file, output_file)
```

```python
# 7. feladat

# Határozzuk meg hogy egy szövegállományban hány nyomtatható karakter van.

def count_printable_characters(input_file):
    """Számolja a nyomtatható karakterek számát egy szövegállományban."""
    with open(input_file, 'r') as file:
        text = file.read()  # Beolvassa a fájl teljes tartalmát
    
    # Nyomtatható karakterek számlálása
    printable_count = sum(1 for char in text if char.isprintable() and char != '\n' and char != '\r')
    
    return printable_count

# Fájl neve
input_file = "input_text_file.txt"  # A bemeneti szövegfájl neve

# Nyomtatható karakterek száma
printable_characters = count_printable_characters(input_file)

print(f"A szövegállományban {printable_characters} nyomtatható karakter található.")
```

```python
# 8. feladat

# Írjunk programot, amely titkosítja a megadott bináris állományt, 
# XOR műveletet használva. A titkosításhoz használt jelszót véletlenszerűen állítsuk elő.

import os
import random
import string

def generate_random_key(length):
    """Generáljon egy véletlenszerű kulcsot, ami megfelel a kívánt hosszúságnak."""
    return ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(length))

def xor_encrypt(input_file, output_file, key):
    """Titkosítja a bináris fájlt XOR művelettel a megadott kulcs segítségével."""
    # A kulcs hosszának beállítása
    key_length = len(key)

    # Megnyitjuk a bemeneti és kimeneti fájlt
    with open(input_file, 'rb') as infile, open(output_file, 'wb') as outfile:
        # A bemeneti fájl tartalmának beolvasása byte-onként
        byte = infile.read(1)
        key_index = 0
        
        while byte:
            # XOR titkosítás
            encrypted_byte = ord(byte) ^ ord(key[key_index])
            
            # Írás a kimeneti fájlba
            outfile.write(bytes([encrypted_byte]))
            
            # A kulcs indexének frissítése
            key_index = (key_index + 1) % key_length

            # Következő byte
            byte = infile.read(1)

def xor_decrypt(input_file, output_file, key):
    """Dekódolja a titkosított fájlt XOR művelettel a megadott kulcs segítségével."""
    # A kulcs hosszának beállítása
    key_length = len(key)

    # Megnyitjuk a bemeneti és kimeneti fájlt
    with open(input_file, 'rb') as infile, open(output_file, 'wb') as outfile:
        # A bemeneti fájl tartalmának beolvasása byte-onként
        byte = infile.read(1)
        key_index = 0
        
        while byte:
            # XOR titkosítás (ugyanaz a művelet titkosításhoz és dekódoláshoz)
            decrypted_byte = ord(byte) ^ ord(key[key_index])
            
            # Írás a kimeneti fájlba
            outfile.write(bytes([decrypted_byte]))
            
            # A kulcs indexének frissítése
            key_index = (key_index + 1) % key_length

            # Következő byte
            byte = infile.read(1)

# Példa használat
input_file = "input_binary_file.bin"  # A titkosítandó bináris fájl
encrypted_file = "encrypted_file.bin"  # Titkosított fájl
decrypted_file = "decrypted_file.bin"  # Visszafejtett fájl

# Véletlenszerű kulcs generálása
key = generate_random_key(16)  # A kulcs hossza legyen 16 karakter

print(f"Használt jelszó: {key}")

# Titkosítás
xor_encrypt(input_file, encrypted_file, key)

# Dekódolás (a kulcs ugyanaz)
xor_decrypt(encrypted_file, decrypted_file, key)

print(f"A titkosított fájl elérhető itt: {encrypted_file}")
print(f"A visszafejtett fájl elérhető itt: {decrypted_file}")
```

```python
# 9. feladat

# Olvassunk be egy tetszőleges karakterláncot a billentyűzetről és 
# határozzuk meg a base64-es alakját.

import base64

def to_base64(input_string):
    # A karakterlánc bájtokká konvertálása
    byte_data = input_string.encode('utf-8')
    
    # Base64 kódolás
    base64_data = base64.b64encode(byte_data)
    
    # A base64 kódolt adat dekódolása, hogy stringet kapjunk vissza
    return base64_data.decode('utf-8')

# Billentyűzetről beolvasott karakterlánc
input_string = input("Adja meg a karakterláncot: ")

# A base64-es kódolás
encoded_string = to_base64(input_string)

print(f"A base64-es kódolt változat: {encoded_string}")
```

```python
# 10. feladat

# Generáljunk véletlenszerűen egy n (n >= 1024) bites számot, alakítsuk 
# át 256-os számrendszerbe, határozzuk meg a base64-es alakját és az eredményt 
# írjuk ki egy állományba. Írjuk meg a visszaalakító műveletsort is: olvassuk 
# ki az állámányból a kódolt szöveget, dekódoljuk, majd a kapott 256-os 
# számrendszerbeli "számot" alakítsuk vissza 10-es számrendszerbe.

import base64
import random

# Függvény, amely egy n bites véletlenszerű számot generál
def generate_random_number(n):
    return random.getrandbits(n)

# Függvény, amely átalakítja az n-bites számot 256-os számrendszerbeli számjegyekre
def to_base256(number):
    base256_digits = []
    while number > 0:
        base256_digits.append(number % 256)
        number = number // 256
    return base256_digits[::-1]  # Fordítva kell, hogy legyen

# Függvény, amely egy számot base64-re kódol
def to_base64(data):
    # Átalakítjuk byte-okba
    byte_data = bytes(data)
    return base64.b64encode(byte_data).decode('utf-8')

# Függvény, amely dekódolja a base64-es kódolást és visszaalakítja 256-os számrendszerbe
def from_base64(encoded_data):
    byte_data = base64.b64decode(encoded_data)
    number = int.from_bytes(byte_data, byteorder='big')
    return to_base256(number)

# Függvény, amely visszaalakítja a 256-os számrendszerbeli számot 10-es számrendszerbe
def base256_to_decimal(base256_data):
    number = 0
    for digit in base256_data:
        number = number * 256 + digit
    return number

# Generálunk egy n-bites véletlenszerű számot
n = 1024  # Például 1024 bites számot generálunk
random_number = generate_random_number(n)

# Átalakítjuk 256-os számrendszerbe
base256_data = to_base256(random_number)

# Kódoljuk base64-re
encoded_base64 = to_base64(base256_data)

# Kiírjuk az eredményt egy fájlba
with open("encoded_data.txt", "w") as file:
    file.write(encoded_base64)

# Most olvassuk vissza a kódolt adatot és dekódoljuk
with open("encoded_data.txt", "r") as file:
    encoded_base64_from_file = file.read()

# Dekódoljuk a base64-es adatot és visszaalakítjuk 256-os számrendszerbeli számra
decoded_base256 = from_base64(encoded_base64_from_file)

# Visszaalakítjuk a 256-os számrendszerbeli számot 10-es számrendszerbe
decimal_number = base256_to_decimal(decoded_base256)

# Eredmény kiírása
print(f"Az eredeti szám 10-es számrendszerbeli alakja: {random_number}")
print(f"A dekódolt szám 10-es számrendszerbeli alakja: {decimal_number}")
```

```python
# 11. feladat

# Titkosítsunk a tanult technikával, az XOR műveletet használva, 
# egy 2Mbnál nagyobb bináris állományt, ahol a kulcsot véletlenszerűen 
# generáljuk. A generált kulcs base64-es alakját írjuk ki egy kulcs.txt 
# szövegállományba, majd a visszafejtéshez a kulcs.txt állományba levő értéket használjuk

import os
import base64
import random

def generate_key(length):
    """Generates a random key of the specified length."""
    return os.urandom(length)

def xor_encrypt(data, key):
    """Encrypts data using XOR with the given key."""
    return bytes([data[i] ^ key[i] for i in range(len(data))])

# Titkosítás
def encrypt_file(input_file, output_file, key_file):
    # Read the input binary file
    with open(input_file, 'rb') as f:
        data = f.read()
    
    # Generate a random key
    key = generate_key(len(data))
    
    # Encrypt the data
    encrypted_data = xor_encrypt(data, key)
    
    # Write the encrypted data to the output file
    with open(output_file, 'wb') as f:
        f.write(encrypted_data)
    
    # Write the key to the key file in Base64 format
    with open(key_file, 'w') as f:
        f.write(base64.b64encode(key).decode('utf-8'))
    
    print(f"Encryption complete. Encrypted file: {output_file}, Key file: {key_file}")

# Visszafejtés
def decrypt_file(encrypted_file, output_file, key_file):
    # Read the encrypted binary file
    with open(encrypted_file, 'rb') as f:
        encrypted_data = f.read()
    
    # Read the key from the key file and decode it from Base64
    with open(key_file, 'r') as f:
        key = base64.b64decode(f.read())
    
    # Decrypt the data
    decrypted_data = xor_encrypt(encrypted_data, key)
    
    # Write the decrypted data to the output file
    with open(output_file, 'wb') as f:
        f.write(decrypted_data)
    
    print(f"Decryption complete. Decrypted file: {output_file}")

# Példa futtatás
input_file = 'input.bin'   # Eredeti bemeneti állomány
encrypted_file = 'encrypted.bin'  # Titkosított kimeneti állomány
decrypted_file = 'decrypted.bin'  # Visszafejtett kimeneti állomány
key_file = 'kulcs.txt'  # Kulcs szövegállomány

# Ellenőrizzük, hogy a bemeneti fájl létezik-e
if not os.path.exists(input_file):
    print(f"A bemeneti fájl ({input_file}) nem található.")
else:
    # Titkosítás
    encrypt_file(input_file, encrypted_file, key_file)
    
    # Visszafejtés
    decrypt_file(encrypted_file, decrypted_file, key_file)

    # Ellenőrizzük, hogy az eredeti és visszafejtett fájl megegyezik-e
    with open(input_file, 'rb') as f1, open(decrypted_file, 'rb') as f2:
        assert f1.read() == f2.read(), "Visszafejtés sikertelen: az állományok nem egyeznek!"
    print("A visszafejtett állomány megegyezik az eredetivel.")
```
