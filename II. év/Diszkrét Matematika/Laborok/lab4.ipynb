```python
# 2. feladat

# Írjunk egy Python függvényt, amely meghatározza a következő 
# számok lánctörtjegyeit: 1.4142135623730951, 2.23606797749979, 3.141592653589793, 
# 2.718281828459045, 1.618033988749895.

import mpmath

# Lánctörtjegyek számítása
def continued_fraction(x, max_iter=100):
    mpmath.mp.dps = 50  # 50 tizedesjegy pontossága
    a = []
    while max_iter > 0:
        # Kiszámoljuk az egész részt
        integer_part = int(mpmath.floor(x))
        a.append(integer_part)
        
        # Számítsuk ki a maradékot
        remainder = x - integer_part
        if remainder == 0:
            break
        
        # Vegyük a reciprokát
        x = 1 / remainder
        max_iter -= 1
    
    return a

# A kért számok
numbers = [1.4142135623730951, 2.23606797749979, 3.141592653589793, 2.718281828459045, 1.618033988749895]

# Eredmények kiírása
for number in numbers:
    fraction = continued_fraction(number)
    print(f"Lánctörtjegyek a számhoz {number}: {fraction}")
```

```python
# 3. feladat

# Írjunk egy Python függvényt, amely meghatározza több, billentyűzetről 
# beolvasott szám, k-ik gyökét, alkalmazva a következőket, ahol az iterációt 
# addig kell végezni, amíg xn+1 nem egyenlő (vagy alig különbözik) xn-nel:
#   x_0 = 1
#   x_{n+1} = ((k-1)·x_{n} + nr/(x_{n}^{k-1}))/k 

def k_th_root(r, k, tolerance=1e-10, max_iter=1000):
    # Kezdeti közelítés
    x_n = 1.0
    
    # Iteráció
    for _ in range(max_iter):
        # Számoljuk ki a következő közelítést
        x_next = ((k-1) * x_n + r / (x_n ** (k-1))) / k
        
        # Ellenőrizzük, hogy elértük-e a kívánt pontosságot
        if abs(x_next - x_n) < tolerance:
            return x_next
        
        # Frissítjük az aktuális közelítést
        x_n = x_next
    
    # Ha túllépjük a max_iter-t, akkor visszaadjuk a legjobb közelítést
    return x_n

# Beolvassuk a számokat és k-t
n = int(input("Hány számot szeretnél beolvasni? "))
    
for _ in range(n):
    r, k = map(int, input("Add meg a számot és a gyököt (r, k): ").split())
    result = k_th_root(r, k)
    print(f"A {r} {k}-ik gyöke: {result}")
```

```python
# 4. feladat

# Írjunk egy Python függvényt, amely meghatározza egy másodfokú egyenlet komplex gyökeit.

import cmath  # cmath modul a komplex számok kezelésére

def complex_roots(a, b, c):
    # A diszkrimináns számítása
    D = b**2 - 4*a*c
    
    # Két komplex gyök kiszámítása
    root1 = (-b + cmath.sqrt(D)) / (2 * a)
    root2 = (-b - cmath.sqrt(D)) / (2 * a)
    
    return root1, root2

# Beolvassuk a felhasználótól a másodfokú egyenlet együtthatóit
a = float(input("Add meg az a együtthatót: "))
b = float(input("Add meg a b együtthatót: "))
c = float(input("Add meg a c együtthatót: "))
    
# Meghatározzuk a komplex gyököket
root1, root2 = complex_roots(a, b, c)
    
print(f"A gyökök: {root1} és {root2}")
```

```python
# 5. feladat

# Írjunk egy-egy Python függvényt, amely meghatározza a komplex 
# számokkal végezhető alapműveletek eredményeit: összeadás, 
# kivonás, szorzás, osztás, abszolutérték, hatványozás. 
# A komplex számokat valós értékpárokként kezeljük.

import math

# Komplex számok összeadása
def add_complex(c1, c2):
    real = c1[0] + c2[0]
    imag = c1[1] + c2[1]
    return (real, imag)

# Komplex számok kivonása
def subtract_complex(c1, c2):
    real = c1[0] - c2[0]
    imag = c1[1] - c2[1]
    return (real, imag)

# Komplex számok szorzása
def multiply_complex(c1, c2):
    real = c1[0]*c2[0] - c1[1]*c2[1]
    imag = c1[0]*c2[1] + c1[1]*c2[0]
    return (real, imag)

# Komplex számok osztása
def divide_complex(c1, c2):
    denominator = c2[0]**2 + c2[1]**2
    real = (c1[0]*c2[0] + c1[1]*c2[1]) / denominator
    imag = (c1[1]*c2[0] - c1[0]*c2[1]) / denominator
    return (real, imag)

# Komplex szám abszolút értéke
def abs_complex(c):
    return math.sqrt(c[0]**2 + c[1]**2)

# Komplex szám hatványozás
def power_complex(c, n):
    result_real, result_imag = c[0], c[1]
    for _ in range(n-1):
        result_real, result_imag = multiply_complex((result_real, result_imag), c)
    return (result_real, result_imag)

c1 = (3, 4)  # 3 + 4i
c2 = (1, 2)  # 1 + 2i

print(f"Összeadás: {add_complex(c1, c2)}")
print(f"Kivonás: {subtract_complex(c1, c2)}")
print(f"Szorzás: {multiply_complex(c1, c2)}")
print(f"Osztás: {divide_complex(c1, c2)}")
print(f"Abszolút érték c1: {abs_complex(c1)}")
print(f"Abszolút érték c2: {abs_complex(c2)}")
print(f"Hatványozás (c1^2): {power_complex(c1, 2)}")
```
