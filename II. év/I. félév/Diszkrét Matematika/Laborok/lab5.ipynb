```python
# 1. feladat

# Írjunk függvényeket, amelyek meghatározzák a következőket:

#   - egy tízes számrendszerbeli szám első számjegyét
#   - egy tízes számrendszerbeli szám számjegyeinek számát
#   - egy tízes számrendszerbeli szám páros számjegyeinek számát
#   - egy tízes számrendszerben megadott szám, b számrendszerbeli első számjegyét
#   - egy tízes számrendszerben megadott szám, b számrendszerbeli számjegyei közül 
#    a k-val egyenlő számjegyek számát.

def first_digit(n):
    # A szám első számjegyét adjuk vissza
    return int(str(abs(n))[0])

def digit_count(n):
    # A szám számjegyeinek számát adjuk vissza
    return len(str(abs(n)))

def even_digit_count(n):
    # A szám páros számjegyeinek számát adjuk vissza
    return sum(1 for digit in str(abs(n)) if int(digit) % 2 == 0)

def first_digit_in_base(n, b):
    # A szám első számjegyét adja vissza a b számrendszerben
    n_in_base_b = ''
    while n > 0:
        n_in_base_b = str(n % b) + n_in_base_b
        n = n // b
    return int(n_in_base_b[0]) if n_in_base_b else 0

def count_digit_in_base(n, b, k):
    # A b számrendszerbeli számjegyek közül hány k-val egyenlő számjegy van
    n_in_base_b = ''
    while n > 0:
        n_in_base_b = str(n % b) + n_in_base_b
        n = n // b
    return n_in_base_b.count(str(k))

num = 123456
base = 2
k = 1

# 1. első számjegy
print(f"Első számjegy: {first_digit(num)}")

# 2. számjegyek száma
print(f"Számjegyek száma: {digit_count(num)}")

# 3. páros számjegyek száma
print(f"Páros számjegyek száma: {even_digit_count(num)}")

# 4. első számjegy b számrendszerben
print(f"Első számjegy {base}-es számrendszerben: {first_digit_in_base(num, base)}")

# 5. k-val egyenlő számjegyek száma b számrendszerben
print(f"A {base}-es számrendszerben {k}-val egyenlő számjegyek száma: {count_digit_in_base(num, base, k)}")
```

```python
# 2. feladat

# Írjunk függvényt, amely

#   - meghatározza egy szám b-es számrendszerbeli számjegyei alapján, a b^k számrendszerbeli alakot
#   - meghatározza egy szám b^k-es számrendszerbeli számjegyei alapján, a b számrendszerbeli alakot
#   - meghatározza egy szám 2-es számrendszerbeli számjegyei alapján, hogy melyik a szám 
#     legnagyobb "számjegye" 2^k számrendszerben
#   - meghatározza egy szám 2^k számrendszerben megadott értéke alapján, hogy hány 1-es számjegy 
#     van a szám 2-es számrendszerbeli alakjában
#   - megvizsgálja, hogy két szám, 2^k számrendszerben megadott értéke alapján melyik számnak 
#     a 2-es számrendszerbeli alakjában van több 1-es.

def to_base_bk_from_base_b(digits, b, k):
    # A szám b-es számrendszerbeli számjegyei alapján visszaállítjuk a számot a b^k számrendszerbe
    num_in_base_b = sum(digit * (b ** i) for i, digit in enumerate(reversed(digits)))
    result = []
    
    # A számot b^k számrendszerbe alakítjuk
    while num_in_base_b > 0:
        result.append(num_in_base_b % (b ** k))
        num_in_base_b = num_in_base_b // (b ** k)
        
    return result[::-1]  # Visszafelé írjuk ki a számjegyeket

def to_base_b_from_base_bk(digits, b, k):
    # A b^k-es számrendszerbeli számjegyeket először b-es számrendszerre konvertáljuk
    num_in_base_bk = sum(digit * (b ** (k * i)) for i, digit in enumerate(reversed(digits)))
    result = []
    
    # A számot b-es számrendszerbe alakítjuk
    while num_in_base_bk > 0:
        result.append(num_in_base_bk % b)
        num_in_base_bk = num_in_base_bk // b
        
    return result[::-1]  # Visszafelé írjuk ki a számjegyeket

def largest_digit_in_base_2k(digits, k):
    # A szám 2-es számrendszerbeli számjegyei alapján meghatározzuk a legnagyobb "számjegyet" 2^k számrendszerben
    num_in_base_2 = sum(digit * (2 ** i) for i, digit in enumerate(reversed(digits)))
    max_digit = 0
    while num_in_base_2 > 0:
        max_digit = max(max_digit, num_in_base_2 % (2 ** k))
        num_in_base_2 = num_in_base_2 // (2 ** k)
    
    return max_digit

def count_ones_in_base_2(digits, k):
    # A szám 2^k számrendszerbeli alakja alapján megszámoljuk a 2-es számrendszerbeli 1-es számjegyeket
    num_in_base_2k = sum(digit * (2 ** (k * i)) for i, digit in enumerate(reversed(digits)))
    binary_representation = bin(num_in_base_2k)[2:]  # Bináris alakba konvertáljuk
    return binary_representation.count('1')  # Megszámoljuk a 1-eseket

def compare_ones_in_base_2(digits1, digits2, k):
    # Két szám 2^k számrendszerbeli értéke alapján összehasonlítjuk a 2-es számrendszerbeli 1-es számjegyeket
    ones_count_1 = count_ones_in_base_2(digits1, k)
    ones_count_2 = count_ones_in_base_2(digits2, k)
    
    if ones_count_1 > ones_count_2:
        return "Az első számnak több 1-es számjegye van."
    elif ones_count_1 < ones_count_2:
        return "A második számnak több 1-es számjegye van."
    else:
        return "Mindkét számnak ugyanannyi 1-es számjegye van."
    
digits1 = [1, 0, 1, 1]  # 2-es számrendszerbeli szám
digits2 = [1, 0, 1]     # 2-es számrendszerbeli szám
k = 2

# 1. b^k számrendszerbeli alak meghatározása
print(f"b^k számrendszerbeli alak: {to_base_bk_from_base_b(digits1, 2, k)}")

# 2. b-es számrendszerbeli alak meghatározása
print(f"b-es számrendszerbeli alak: {to_base_b_from_base_bk(digits1, 2, k)}")

# 3. Legnagyobb számjegy 2^k számrendszerben
print(f"Legnagyobb számjegy 2^k számrendszerben: {largest_digit_in_base_2k(digits1, k)}")

# 4. 1-es számjegyek száma 2-es számrendszerbeli alakban
print(f"1-es számjegyek száma 2-es számrendszerbeli alakban: {count_ones_in_base_2(digits1, k)}")

# 5. Összehasonlítás
print(compare_ones_in_base_2(digits1, digits2, k))
```

```python
# 3. feladat

# Adjuk meg a számrendszerek közötti átalakító függvények rekurzív változatát.

def to_base_bk_from_base_b_recursive(n, b, k):
    if n == 0:
        return []
    else:
        # Rekurzív hívás: a számot az aktuális b^k számrendszerbeli helyen elmentjük, majd a maradékot rekurzióval kezeljük
        return to_base_bk_from_base_b_recursive(n // (b ** k), b, k) + [n % (b ** k)]
    
def to_base_b_from_base_bk_recursive(digits, b, k):
    if len(digits) == 0:
        return 0
    else:
        # Rekurzív hívás: az első számjegy szorzása a megfelelő b^k hatvánnyal, és hozzáadjuk a következő részletet
        return digits[0] * (b ** (k * (len(digits) - 1))) + to_base_b_from_base_bk_recursive(digits[1:], b, k)
    
    
# Bármelyik függvény használható az alábbi módon:
n = 15  # 10-es számrendszer
b = 2   # Bázis
k = 2   # B^k számrendszer

# b-es számrendszerből b^k-es számrendszerbe
print(to_base_bk_from_base_b_recursive(n, b, k))

# b^k-es számrendszerből b-es számrendszerbe
digits_bk = [3, 1, 0, 4]
print(to_base_b_from_base_bk_recursive(digits_bk, b, k))
```

```python
# 4. feladat

# Adott n szám esetében határozzuk meg azokat a Fibonacci számokat, 
# amelyek összegeként felírható a szám, majd határozzuk meg a szám 
# Fibonacci számrendszerben felírt alakját.

def fibonacci_sequence(limit):
    fib = [0, 1]
    while fib[-1] <= limit:
        fib.append(fib[-1] + fib[-2])
    return fib[2:]  # A Fibonacci számrendszer a 0 és 1 számokat nem tartalmazza

def fibonacci_representation(n):
    fib_sequence = fibonacci_sequence(n)
    representation = []
    
    # Kezdjük a legnagyobb Fibonacci számmal és dolgozzunk visszafelé
    for fib_num in reversed(fib_sequence):
        if fib_num <= n:
            representation.append(1)
            n -= fib_num
        else:
            representation.append(0)
    
    return representation

def fibonacci_number_system(n):
    fib_rep = fibonacci_representation(n)
    fib_sequence = fibonacci_sequence(n)
    
    # Nyomtassuk ki a Fibonacci számokat és azokhoz tartozó 0/1 értékeket
    result = []
    for i, val in enumerate(fib_rep):
        if val == 1:
            result.append(str(fib_sequence[i]))
    
    # Az így kapott Fibonacci számok összegeként a szám felírható
    return ' + '.join(result)

# Példa használat
n = 100
fib_rep = fibonacci_representation(n)
fib_rep_str = fibonacci_number_system(n)

print(f"A {n} Fibonacci számrendszerbeli alakja: {fib_rep_str}")
```

```python
# 5. feladat

# Olvassunk be stringként egy bináris számsorozatot. 
# Ha feltételezzük, hogy a bináris számsorozat valemely Fibonacci 
# számrendszerbe felírt számnak felel meg, akkor határozzuk meg a számot.

def fibonacci_sequence(limit):
    # Generáljunk Fibonacci sorozatot, amíg el nem érjük a limitet
    fib = [0, 1]
    while fib[-1] <= limit:
        fib.append(fib[-1] + fib[-2])
    return fib[2:]  # A Fibonacci sorozat első két elemét elhagyjuk (0, 1)

def fibonacci_from_binary(binary_str):
    # A bináris string alapján meghatározzuk a számot
    n = len(binary_str)
    
    # Készítsük el a szükséges Fibonacci számokat
    fib_sequence = fibonacci_sequence(n)
    
    # Számoljuk össze a Fibonacci számokat, ahol a bináris számjegy 1
    total = 0
    for i in range(n):
        if binary_str[i] == '1':
            total += fib_sequence[i]
    
    return total

# Példa használat
binary_str = input("Adj meg egy bináris számot (Fibonacci számrendszer): ")
result = fibonacci_from_binary(binary_str)
print(f"A szám értéke: {result}")
```

```python
# 6. feladat

# Határozzuk meg az n-ik Fibonacci szám első és utolsó számjegyét, ahol (n > 10000). 

import math

def fibonacci_first_digit(n):
    if n == 0:
        return 0
    phi = (1 + math.sqrt(5)) / 2  # Aranymetszés
    log_phi = math.log10(phi)
    log_sqrt_5 = math.log10(math.sqrt(5))
    
    # A Fibonacci szám logaritmusának számítása
    log_fib_n = n * log_phi - log_sqrt_5
    fractional_part = log_fib_n - int(log_fib_n)
    
    # Az első számjegy kiszámítása
    first_digit = int(10 ** fractional_part)
    return first_digit

# Példa használat
n = 10000
print(f"A {n}-ik Fibonacci szám első számjegye: {fibonacci_first_digit(n)}")

def fibonacci_last_digit(n):
    # Csak az utolsó számjegy számítása modulo 10
    if n == 0:
        return 0
    elif n == 1:
        return 1
    
    prev, curr = 0, 1
    for _ in range(2, n + 1):
        prev, curr = curr, (prev + curr) % 10
    return curr

# Példa használat
n = 10000
print(f"A {n}-ik Fibonacci szám utolsó számjegye: {fibonacci_last_digit(n)}")
```

```python
# 7. feladat

# Adott n szám esetében határozzuk meg a faktoriális számrendszerbeli számjegyeket.

def factorial_base(n):
    # A faktoriális számrendszer számjegyének listája
    result = []
    
    # Keresés kezdete
    i = 1
    while i <= n:
        i *= len(result) + 1  # Keresd a legnagyobb faktoriálist, amely kisebb vagy egyenlő n-nel
    
    # Faktoriális számrendszer kifejezése
    for i in range(len(result), -1, -1):
        fact = 1
        for j in range(1, i+1):
            fact *= j
        digit = n // fact
        result.append(digit)
        n = n % fact
    
    return result

# Példa használat
n = 463
result = factorial_base(n)
print(f"A {n} faktoriális számrendszerbeli kifejezése: {result}")
```

```python
# 8. feladat

# Olvassunk be stringként egy számsorozatot, amely feltételezhetően 
# egy szám faktoriális számrendszerbeli számjegyeit jelentik, majd 
# határozzuk, meg a tízes számrendszerbeli számot.

def factorial_to_decimal(fact_digits):
    """
    Átalakít egy faktoriális számrendszerbeli számot tízes számrendszerbe.
    
    fact_digits: A számjegyek listája (a faktoriális számrendszer számjegyei)
    
    Visszatérési érték: A tízes számrendszerbeli szám.
    """
    decimal_value = 0
    # A faktoriális számrendszer számjegyei a legnagyobb faktoriálistól kezdődnek, tehát:
    # a_0 a 0!, a_1 a 1!, a_2 a 2!, stb.
    for i, digit in enumerate(reversed(fact_digits)):
        factorial = 1
        for j in range(1, i+1):
            factorial *= j
        decimal_value += digit * factorial
    return decimal_value

# Példa használat
input_str = input("Adjon meg egy faktoriális számrendszerbeli számot (pl. 3411): ")

# A bemeneti stringet számjegyek listájává alakítjuk
fact_digits = [int(digit) for digit in input_str]

# Az átalakított tízes számrendszerbeli érték
decimal_value = factorial_to_decimal(fact_digits)

print(f"A szám a tízes számrendszerben: {decimal_value}")
```
